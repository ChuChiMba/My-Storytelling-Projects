```python
plt.figure(figsize=(4, 2))
explode= (0.05, 0.05, 0.05, 0.05, 0.05)
df1.groupby(['Region']).sum().plot(kind='pie', y='Average Consumption', autopct='%1.0f%%', title='Regional Consumption Rates of Coca-Cola Beverages', explode=explode)
```

```python
plt.figure(figsize=(4, 2))
Mylabels= ('Eurasia & Africa', 'Europe', 'Latin America', 'North America', 'Asia-Pacific')
explode= (0.05, 0.05, 0.05, 0.05, 0.05)
df.plot(kind='pie', y='Average', autopct='%1.0f%%', title='Unit Case Volume Distribution of Coca-Cola Beverages', labels=Mylabels, explode=explode)
```
